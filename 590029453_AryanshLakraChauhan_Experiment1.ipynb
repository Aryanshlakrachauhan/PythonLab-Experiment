4. IDLE (Interactive Development and Learning Environment) is a built-in editor that comes with Python.

5. IDLE can be launched from the Start Menu (Windows) or Applications folder (macOS).

6. Interactive mode executes Python code line by line and shows the output immediately.

7. Interactive mode is mainly used for testing, learning, and quick calculations and uses the >>> prompt.

8. Scripting mode allows writing Python programs in a file with a .py extension.

9. In scripting mode, the entire program is executed at once and is used for developing larger programs.

```python
# Q5.Declare these variables (x, y and z) as integers. Assign a value of 9 to x, Assign a value of 7 to y, perform addition, multiplication, division and subtraction on these two variables and Print out the result.
x = 9                    # first number
y = 7                    # second number

print(x + y)             # addition
print(x * y)             # multiplication
print(x / y)             # division
print(x - y)             # subtraction
```

```python
# Q6. Write a program to compute the length of the hypotenuse (c) of a right triangle using Pythagoras theorem.

a = 3                    # side a
b = 4                    # side b

c = (a*a + b*b) ** 0.5   # hypotenuse formula
print(c)
```

```python
# Q7. Write a program to find simple interest
p = 1000                 # principal
r = 5                    # rate
t = 2                    # time
si = (p * r * t) / 100   # simple interest formula
print(si)
```

```python
# Q8. Write a program to find area of triangle when length of sides are given.
a = 3
b = 4
c = 5
s = (a + b + c) / 2      # semi-perimeter
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
print(area)
```

```python
# Q9. Write a program to convert given seconds into hours, minutes and remaining seconds.

seconds = int(input("enter the seconds"))

hours = seconds // 3600
seconds = seconds % 3600
minutes = seconds // 60
seconds = seconds % 60

print(hours, minutes, seconds)
```

```python
# Q10. Write a program to swap two numbers without taking additional variable.
a = 5
b = 10
a = a + b
b = a - b
a = a - b
print(a, b)
```

```python
# Q11. Write a program to find sum of first n natural numbers.
n = 10
sum_n = n * (n + 1) // 2  # formula
print(sum_n)
```

```python
# Q12. Write a program to print truth table for bitwise operators (&, | and ^ operators)
print("A B A&B A|B A^B")
for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, A & B, A | B, A ^ B)
```

```python
# Q13. Write a program to find left shift and right shift values of a given number.
num = 5
print(num << 1)          # left shift
print(num >> 1)          # right shift
```

```python
#14 Using membership operator find whether a given number is in sequence (10,20,56,78,89)
sequence = (10, 20, 56, 78, 89)
number_to_find = 56
if number_to_find in sequence:
  print(f"{number_to_find} is in the sequence.")
else:
  print(f"{number_to_find} is not in the sequence.")
another_number = 99
if another_number in sequence:
  print(f"{another_number} is in the sequence.")
else:
  print(f"{another_number} is not in the sequence.")
```
