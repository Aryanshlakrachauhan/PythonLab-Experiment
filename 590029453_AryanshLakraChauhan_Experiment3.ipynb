```python
# Q1: Find factorial of a given number

n = int(input("Enter number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i   # multiply

print("Factorial =", fact)
```

```python
# Q2: Find whether the given number is Armstrong number
num = int(input("Enter number: "))
temp = num
digits = len(str(num))
arm = 0

while temp > 0:
    d = temp % 10
    arm += d ** digits
    temp //= 10

if arm == num:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

```python
# Q3: Print Fibonacci series up to given term.

n = int(input("Enter terms: "))
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Q4: Write a program to find if given number is prime number or not

num = int(input("Enter number: "))
flag = 0

if num <= 1:
    flag = 1
else:
    for i in range(2, num):
        if num % i == 0:
            flag = 1
            break

if flag == 0:
    print("Prime")
else:
    print("Not Prime")
```

```python
# Q5: Check whether given number is palindrome or not

num = int(input("Enter number: "))
temp = num
rev = 0

while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Q6: Write a program to print sum of digits

num = int(input("Enter number: "))
s = 0

while num > 0:
    s += num % 10
    num //= 10

print("Sum =", s)
```

```python
# Q7: Count and print all numbers divisible by 5 or 7 between 1 to

count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nCount =", count)
```

```python
# Q8: Convert all lower cases to upper case in a string.

s = input("Enter string: ")
print("Uppercase:", s.upper())
```

```python
# Q9: Print the table for a given number

num = int(input("Enter number: "))

for i in range(1, 11):
    print(num, "*", i, "=", num * i)
```

```python
# Q10: Write a program to print the following pattern

n = 5

for i in range(n):
    for j in range(1, n - i + 1):
        print(j, end="")
    for j in range(i):
        print("*", end="")
    for j in range(i):
        print("*", end="")
    for j in range(n - i, 0, -1):
        print(j, end="")
    print()
```

```python
# Q11: Write a program to print the sum of the following series

n = int(input("Enter n: "))
sum_series = 0

for i in range(1, n + 1):
    sum_series += 1 / i

print("Sum =", sum_series)
```
